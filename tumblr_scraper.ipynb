```python
import pytumblr
import pandas as pd
import calendar
import time
from unidecode import unidecode
from datetime import datetime
```

```python
# Enter your Tumblr API key here:
tumblr = pytumblr.TumblrRestClient(
    'KimMDjUgZHOJrvkpxmhPfiA0GqnIDIFC1YPgxoB9Z7vUSgXgke'
)

# Set search parameters:
tag = 'mental health'
filter = 'raw'

# Set the date range (epoch timestamps for 01-01-2023 and 01-01-2024)
start_date = int(datetime(2020, 1, 1).timestamp())
end_date = int(datetime(2020, 1, 31).timestamp())

# Initialize a DataFrame to store the results
columns = ["timeStamp", "URL", "blogName", "title", "tags", "body", "reblogs", "notes", "replies", "shares", "likes"]
data = []

print(f"Searching Tumblr for posts tagged '{tag}' from 01-01-2023 to 01-01-2024...")

batch_count = 1
row_count = 0
before = end_date  # Start scraping before the end_date

# Loop to fetch posts in batches (1000-row limit)
while row_count < 1000:
    # Fetch posts tagged with the specified tag and within the time range
    searchResults = tumblr.tagged(tag, filter=filter, before=before)

    if not searchResults:
        print("No more results found.")
        break

    print(f"Batch {batch_count}: Retrieved {len(searchResults)} results.")

    for post in searchResults:
        # Extract post information
        timestamp = post.get('timestamp', 0)
        if timestamp < start_date:  # Stop if we reach posts before the start_date
            print("Reached the start date.")
            break

        blog_name = post.get('blog_name', 'N/A')
        date = post.get('date', 'N/A')
        url = post.get('post_url', 'N/A')
        title = post.get('title', "Couldn't Get Title")
        tags = post.get('tags', [])
        body = post.get('body', "Couldn't Get Post Body")
        reblogs = post.get('reblogs_count', 0)
        notes = post.get('note_count', 0)
        replies = post.get('replies_count', 0)
        shares = post.get('shared_count', 0)
        likes = post.get('likes', 0)

        # Append the data as a new row
        data.append([date, unidecode(url), blog_name, title, tags, body, reblogs, notes, replies, shares, likes])
        row_count += 1

        # Stop if the row limit is reached
        if row_count >= 1000:
            break

    # Update 'before' timestamp to fetch older posts
    if len(searchResults) > 0:
        before = searchResults[-1]['timestamp']
    else:
        break  # Stop if there are no more results

    batch_count += 1

# Convert the data to a DataFrame
df = pd.DataFrame(data, columns=columns)

# Export the DataFrame to a CSV file
outputPath = ''  # Specify your output directory
filePath = outputPath + f'TumblrSearch-{tag}-2020.csv'
df.to_csv(filePath, index=False, encoding='utf-8')

print(f"Finished! Data saved to {filePath}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re

nltk.download('punkt_tab')
nltk.download('stopwords')

def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower()

    text = re.sub(r'[^a-z\s]', '', text)

    tokens = word_tokenize(text)

    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]

    clean_text = ' '.join(tokens)

    return clean_text
```
